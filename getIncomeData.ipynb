<h1>Get Income Data for Each City from Bestplaces.net</h1>

```python
cityDf=pd.read_csv("20kup.csv")
```

```python
def findIncome(city):
    urlCity=city.replace(" ","_")
    cityLength=len(city)
    r = requests.get("https://www.bestplaces.net/economy/city/indiana/{}".format(urlCity))
    text=r.text
    searchString="The average income of a {} resident is $".format(city)
    startTargetPos=text.find(searchString)+38+cityLength
    endTargetPos=startTargetPos+7
    target=text[startTargetPos:endTargetPos].strip()
    target=target.replace(",","")
    return int(target)
```

```python
for index,row in cityDf.iterrows():
    city=row['NAME2']
    cityDf.loc[index,'AveIncome']=findIncome(city)
```
